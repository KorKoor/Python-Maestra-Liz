```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/KorKoor/Python-Maestra-Liz/main/NetFlix.csv', sep=',')
df.head()
```

```python
df.dtypes
```

```python
print (df['date_added'].head())
```

```python
#Intentamos Convertir
try:
  df['date_added'] = pd.to_datetime(df['date_added'])
except ValueError:
  #Si falla hacer esto
  df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')
```

```python
print(df['date_added'].dtype)
```

```python
#Primeras filas
print(df['date_added'].head())
#Si hay valores NaT, indentificarlos
print(df['date_added'].isna().sum())
```

```python
#Definir un mapeo de ratings coon Un diccionario
rating_map = {
    'G': 1,
    'PG': 2,
    'PG-13': 3,
    'R': 4,
    'NC-17': 5,
    'TV-Y': 1,
    'TV-Y7': 2,
    'TV-G': 1,
    'TV-PG': 2,
    'TV-14': 3,
    'TV-MA': 4
}
```

```python
#Crear una nueva columna
df['rating_num'] = df['rating'].map(rating_map)
```

```python
#Para manejar cualquier rating que no este en el diccionario, puedes usar un valor predeterminado
#Mostrar lass primeras filas
print(df[['rating', 'rating_num']].head())
#Obtener estadisticas
print(df['rating_num'].describe())
```
