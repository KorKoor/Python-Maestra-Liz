#Graficacion

```python
import matplotlib.pyplot as plt
import numpy as np
```

#Datos

```python
x = np.linspace(0,10,100)
y = np.sin(x)
```

#Graficos de Lineas

```python
plt.figure(figsize=(10,4))
plt.plot(x,y)
plt.title('Gráfico de Lineas')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

# Gráfico de Barras


```python
categorias = ['A', 'B', 'C', 'D']
valores = [3, 7, 2, 5]

plt.figure(figsize=(10, 4))
plt.bar(categorias, valores)
plt.title('Gráfico de Barras')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

#Gráfico de Dispersión

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(10, 4))
plt.scatter(x, y)
plt.title('Gráfico de Dispersión')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

# Histograma

```python
datos = np.random.randn(1000)

plt.figure(figsize=(10, 4))
plt.hist(datos, bins=30)
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```
